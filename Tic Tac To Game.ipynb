```python
def display_pattern(value):
    print("\n")
    print("\t    |    |    ")
    print("\t  {} |  {} |  {}".format(value[0], value[1], value[2]))
    print("\t______________")
    print("\t    |    |    ")
    print("\t  {} |  {} |  {}".format(value[3], value[4], value[5]))
    print("\t______________")
    print("\t    |    |    ")
    print("\t  {} |  {} |  {}".format(value[6], value[7], value[8]))
    print("\t    |    |    ")

def getnum(player, position, pattern):
    b = "n"
    a = ""
    for i in range(position):
        a = b + a

    pattern[position] = player

    while len(pattern) != 9:
        pattern.append(b)

    pattern_str = "".join(pattern)
    display_pattern(pattern_str)

def check_winner():
    winning_combinations = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]

    if xpat in winning_combinations:
        print("Game ends and X wins")
        return True
    elif opat in winning_combinations:
        print("Game ends and O wins")
        return True
    elif len(all_value) == 9:
        print("Game ends in a draw")
        return True

    return False

def inp(player):
    position = int(input(f"Enter the {player} position (0-8): "))
    if position in all_value:
        print("Already in position")
    else:
        all_value.append(position)
        if player == "X":
            xpat.append(position)
        else:
            opat.append(position)
        getnum(player, position, current_pattern)

d = [0, 1, 2]
c = [3, 4, 5]
v = [6, 7, 8]
u = [0, 3, 6]
r = [1, 4, 7]
j = [2, 5, 8]
t = [2, 4, 6]
q = [0, 4, 8]

all_value = []
xpat = []
opat = []
current_pattern = ["0", "1", "2", "3", "4", "5", "6", "7", "8"]

def main():
    for i in range(9):
        display_pattern("012345678")
        inp("X")
        if check_winner():
            break
        inp("O")
        if check_winner():
            break
        
main()

```
